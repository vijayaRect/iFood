#Transforming Columns

It is too difficult to handle analysis for 37 columns , let's check for the possible ways to cut down it without affecting the data

1. Can be clubbed the 5 marital status columns to one.
2. Similarly with 5 education columns can be migrated as one column

Hence we can reduce 12 columns and will be helpful for further analysis.

```python
df['marital_Divorced']=df['marital_Divorced'].astype(str).replace({'0':''})
df['marital_Married']=df['marital_Married'].astype(str).replace({'1':'2','0':''})
df['marital_Single']=df['marital_Single'].astype(str).replace({'1':'3','0':''})
df['marital_Together']=df['marital_Together'].astype(str).replace({'1':'4','0':''})
df['marital_Widow']=df['marital_Widow'].astype(str).replace({'1':'5','0':''})

df['marital_status']=df['marital_Widow']+df['marital_Together']+df['marital_Single']+df['marital_Married']+df['marital_Divorced']
df['marital_status']=df['marital_status'].map({'1':'Divorced', '2':'Married','3':'Single','4':'Together','5':'Widow'})
```

```python
df['education_2n Cycle']=df['education_2n Cycle'].astype(str).replace({'0':''})
df['education_Basic']=df['education_Basic'].astype(str).replace({'1':'2','0':''})
df['education_Graduation']=df['education_Graduation'].astype(str).replace({'1':'3','0':''})
df['education_Master']=df['education_Master'].astype(str).replace({'1':'4','0':''})
df['education_PhD']=df['education_PhD'].astype(str).replace({'1':'5','0':''})

df['education']=df['education_2n Cycle']+df['education_Basic']+df['education_Graduation']+df['education_Master']+df['education_PhD']
df['education']=df['education'].map({'1':'2n Cycle','2':'Basic','3':'Graduation','4':'Master','5':'PhD'})
```

```python
df.drop(['marital_Divorced', 'marital_Married','marital_Single','marital_Together','marital_Widow','education_2n Cycle','education_Basic','education_Graduation',
         'education_Master','education_PhD'], axis=1, inplace=True)
```

```python
selected_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
X_Names= ['Wines', 'Fruits', 'MeatProducts', 'FishProducts', 'SweetProducts', 'GoldProds']

totalProductsList=[]
[ totalProductsList.append(df[column].sum())  for column in selected_columns]

adults_list = [df.loc[(df['Kidhome'] == 0) & (df['Teenhome'] == 0)][column].sum() for column in selected_columns]
teen_list = [df.loc[(df['Teenhome'] > 0) & (df['Kidhome'] == 0)][column].sum() for column in selected_columns]
total_kids_list = [df.loc[(df['Teenhome'] > 0) & (df['Kidhome'] > 0)][column].sum() for column in selected_columns]
kids_list = [df.loc[(df['Kidhome'] > 0) & (df['Teenhome'] == 0)][column].sum() for column in selected_columns]

indices = np.arange(len(selected_columns))
bar_width = 0.2
fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(indices - 1.5 * bar_width, adults_list, bar_width, label='Adults only')
bars2 = ax.bar(indices - 0.5 * bar_width, teen_list, bar_width, label='Teens Only', color='orange')
bars3 = ax.bar(indices + 0.5 * bar_width, total_kids_list, bar_width, label='Kids & Teens', color='green')
bars4 = ax.bar(indices + 1.5 * bar_width, kids_list, bar_width, label='Kids Only', color='red')

ax.set_xlabel('Products')
ax.set_ylabel('Total Amount Spent')
ax.set_title('Amount Spent for individual Products group by Kids & Teens ')
ax.set_xticks(indices)
ax.set_xticklabels(X_Names)
ax.tick_params(axis='x', rotation=90)  
ax.legend()
for bars, group_list in zip([bars1, bars2, bars3, bars4], [adults_list, teen_list, total_kids_list, kids_list]):
    i=0
    for bar, group_value in zip(bars, group_list):
        height = bar.get_height()
        total=totalProductsList        
        percentage = (group_value / total[i]) * 100
        ax.annotate(f'{percentage:.1f}%', 
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=7)  
        i=i+1
plt.show()
```
